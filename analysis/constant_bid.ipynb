```python
from load_df import load_df
import pandas as pd
dfx = load_df("data/x_train.csv")
```

```python
def is_constant_bid(df_obs:pd.DataFrame):
    obs_id = df_obs["obs_id"].values[0]
    boolean = (len(df_obs["bid"].unique())==1) | (len(df_obs["ask"].unique())==1)
    df = pd.DataFrame({"obs_id":[obs_id],
                       "value":[boolean]})
    return df

grouped = dfx.groupby("obs_id").apply(lambda df_obs: is_constant_bid(df_obs))
grouped.set_index("obs_id",inplace=True)
```
