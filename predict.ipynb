```python
import keras
loaded_model = keras.saving.load_model("pretrained.keras")
```

```python
CATEGORY_COLUMNS = ["venue","action","trade","side"]
SCALAR_COLUMNS = ["price","bid","ask","bid_size","ask_size","flux"]
INDEX_COLUMN= "obs_id"
```

```python
import pandas as pd
testx_df = pd.read_csv("data/test/x_test.csv")
from encoder import Encoder
encoder = Encoder(category_colums=CATEGORY_COLUMNS)
testx_df = encoder.apply(testx_df)
CATEGORY_COLUMNS = encoder.new_columns

from feature_engineer import FeatureEngineer
feature_engineer = FeatureEngineer(category_columns=CATEGORY_COLUMNS,
                                   scalar_columns=SCALAR_COLUMNS,
                                   index_column=INDEX_COLUMN)
```

```python
from splitter import Splitter
splitter_x = Splitter(df=testx_df,
                      batch_size=20000,
                      test_size=0,
                      index_column=INDEX_COLUMN)
```

```python
from preparator import Preparator
preparator = Preparator()

output = []

while not splitter_x.end:
    df = splitter_x.get_current_slice()
    df = feature_engineer.apply(df)
    testx_array = preparator.input_df_to_array(df,index_column=INDEX_COLUMN)
    prediction = loaded_model.predict(testx_array)
    output.extend([max(enumerate(row), key=lambda x: x[1])[0] for row in prediction])
    splitter_x.shift()
```

```python

import pandas as pd
output_df = pd.DataFrame({"obs_id":list(testx_df["obs_id"].unique()),"eq_code_cat":output})
output_df.to_csv("data/submission_feature_engineering.csv",index=False)
```
